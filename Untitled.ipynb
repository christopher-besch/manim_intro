```python
%%manim Example

def function(x):
    return (x+2)*x*(x-2)/2

class Example(Scene):
    def construct(self):
        ax = Axes(x_range=(-3, 3), y_range=(-3, 3))
        self.play(Create(ax, run_time=3))
        curve = ax.plot(function, color=RED)
        self.play(Create(curve, run_time=2))
        area = ax.get_area(curve, x_range=(-2, 0))
        self.play(FadeIn(area))
        self.play(Indicate(area))
        self.wait(2)
```
